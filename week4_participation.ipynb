```python
import pandas as pd

df = pd.read_csv('/content/parsed.csv')
display(df.head())
```

```python
japan_mb_earthquakes = df[(df['parsed_place'].str.contains('Japan')) & (df['magType'] == 'mb')]
percentile_95 = japan_mb_earthquakes['mag'].quantile(0.95)
indonesia_earthquakes = df[df['parsed_place'].str.contains('Indonesia')]
tsunami_count_indonesia = indonesia_earthquakes[indonesia_earthquakes['tsunami'] == 1].shape[0]
total_earthquakes_indonesia = indonesia_earthquakes.shape[0]
tsunami_percentage_indonesia = (tsunami_count_indonesia / total_earthquakes_indonesia) * 100 if total_earthquakes_indonesia > 0 else 0
display(tsunami_percentage_indonesia)
```

```python
nevada_earthquakes = df[df['parsed_place'].str.contains('Nevada')]
display(nevada_earthquakes['mag'].describe())
```

```python
ring_of_fire_locations = [
    'Bolivia', 'Chile', 'Ecuador', 'Peru', 'Costa Rica', 'Guatemala',
    'Mexico', 'Japan', 'Philippines', 'Indonesia', 'New Zealand',
    'Antarctica', 'Canada', 'Fiji', 'Alaska', 'Washington', 'California',
    'Russia', 'Taiwan', 'Tonga', 'Kermadec Islands'
]
```

```python
df['ring_of_fire'] = df['parsed_place'].isin(ring_of_fire_locations)
display(df.head())
```

```python
ring_of_fire_counts = df['ring_of_fire'].value_counts()
display(ring_of_fire_counts)
```

```python
ring_of_fire_tsunami_count = df[(df['ring_of_fire'] == True) & (df['tsunami'] == 1)].shape[0]
display(ring_of_fire_tsunami_count)
```

```python
print(f"95th percentile of earthquake magnitude in Japan (mb type): {percentile_95}")
print(f"Percentage of earthquakes in Indonesia coupled with tsunamis: {tsunami_percentage_indonesia:.2f}%")
print("Summary statistics for earthquake magnitudes in Nevada:")
display(nevada_earthquakes['mag'].describe())
print("Earthquake counts in and out of Ring of Fire locations:")
display(ring_of_fire_counts)
print(f"Tsunami count along the Ring of Fire: {ring_of_fire_tsunami_count}")
```
